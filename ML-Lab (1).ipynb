```python
df = pd.read_csv("used_cars_data.csv")

fuel_cou = {}

for fuel in df['Fuel_Type']:
    if fuel in fuel_cou:
        fuel_cou[fuel] += 1
    else:
        fuel_cou[fuel] = 1

fuel_types = list(fuel_cou.keys())
counts = list(fuel_cou.values())

plt.figure(figsize=(8, 5))
plt.bar(fuel_types, counts, color='skyblue')
plt.xlabel("Fuel Type")
plt.ylabel("Number of Cars")
plt.title("Car Count by Fuel Type")
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
count = data['Transmission'].value_counts()

labels = count.index

color = ['cyan','blue']

plt.pie(count,labels=labels,colors=color)
plt.title("Car Count by Fuel Type")
plt.axis('equal')
plt.show()
```

```python
fuel_types = data['Fuel_Type'].dropna().unique()

plt.figure(figsize=(8, 6))

for fuel in fuel_types:
    prices = data[data['Fuel_Type'] == fuel]['Price'].dropna()
    plt.hist(prices, bins=15, alpha=0.5, label=fuel, edgecolor='black')

plt.title('Price Distribution by Car count')
plt.xlabel('Price')
plt.ylabel('Car Count')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Kilometers Driven vs Price
plt.scatter(data=data, x='Kilometers_Driven', y='Price')
plt.title('Kilometers Driven vs Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

```python
# Average Price by Fuel Type
fuel_avg_price = data.groupby('Fuel_Type')['Price'].mean()

plt.bar(fuel_avg_price.index, fuel_avg_price.values, color='cyan')
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.grid(axis='y')
plt.show()
```
